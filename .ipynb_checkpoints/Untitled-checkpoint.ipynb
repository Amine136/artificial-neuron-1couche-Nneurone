```python
from sklearn.datasets import make_circles
```

```python
X, y = make_circles(n_samples=1000, noise=0.1, factor=0.3, random_state=0)
X = X.T
y = y.reshape((1, y.shape[0]))
```

```python
Experience = artificiel_neuronS(X, y, n, learning_rate, i_max, test)
```

```python
p = Experience['parametres']
```

```python
def plot_decision_boundary(X, y, W1, W2, b1, b2):
    # Crée une grille de points pour évaluer la frontière de décision
    x_min, x_max = X[0].min() - 1, X[0].max() + 1
    y_min, y_max = X[1].min() - 1, X[1].max() + 1
    xx, yy = np.meshgrid(np.arange(x_min, x_max, 0.1),
                         np.arange(y_min, y_max, 0.1))
    
    # Prépare les données pour évaluation
    Z = predict(np.c_[xx.ravel(), yy.ravel()], W1, W2, b1, b2)
    Z = Z.reshape(xx.shape)
    
    # Trace les contours de la frontière de décision
    plt.contourf(xx, yy, Z, alpha=0.8)
    
    # Trace les points de données
    plt.scatter(X[:, 0], X[:, 1], c=y.flatten(), edgecolors='k')
    plt.xlabel('X1')
    plt.ylabel('X2')
    plt.title('Decision Boundary')
    plt.show()
```

```python
def predict(X, W1, W2, b1, b2):
    # Propagation avant
    Z1 = np.dot(X, W1) + b1
    A1 = np.tanh(Z1)
    Z2 = np.dot(A1, W2) + b2
    A2 = 1 / (1 + np.exp(-Z2))
    
    # Prédiction des étiquettes
    predictions = np.round(A2)
    
    return predictions

```

```python
plot_decision_boundary(X, y, p['W1'], p['W2'], p['b1'], p['b2'])
```
